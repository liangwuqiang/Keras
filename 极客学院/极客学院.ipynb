```python
url = 'http://wiki.jikexueyuan.com/project/tensorflow-zh/'  # 极客学院
```

```python
from urllib import request
req = request.Request(url)
res = request.urlopen(req)
html = res.read().decode('utf-8')
print(html)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
alist = soup.select('.detail-nav')[0].select('a')
alist[0:5]
```

```python
items = []
for item in alist:
    if  '.html' in item.get('href'):
        items.append(item.get('href').lstrip('../../'))
items[:5]
```

```python
import hashlib
def md5(name):
    """ 将字符串转成哈希码 """
    if not isinstance(name, str):
        name = str(name)
    md5 = hashlib.md5()
    md5.update(name.encode('utf-8'))
    return md5.hexdigest()
```

```python
from bs4 import BeautifulSoup
import re
import os
pattern = '<img .*?src=\"(.*?)\"'
re_image = re.compile(pattern)
index = 1
for item in items:
    res = request.urlopen(item)
    html = res.read().decode('utf-8')
    
    soup = BeautifulSoup(html, 'html.parser')
    content = soup.select('.markdown-body')[0]
#     print(content)
#     print(type(content))
#     break
    title = soup.select('.markdown-body')[0].h1.text.strip()
    
    content = str(content)

    for image_link in re_image.findall(content):
        if not os.path.exists('images'):
            os.mkdir('images')
        image_link2 = 'http://wiki.jikexueyuan.com/' + image_link
        filename = 'images/' + md5(image_link2) + os.path.splitext(image_link2)[-1]
        try:
            request.urlretrieve(image_link2, filename)
            print('下载完成', filename)
        except Exception as e:
            print('图片出错', e)
        else:
            content = content.replace(image_link, filename)
    print('已完成--------')
    
    html_template = """<!DOCTYPE html>
    <html><head><meta charset="UTF-8">
    </head><body>
    <p><a href="{origin}">原文链接</a></p>
        {content}
    </body></html>"""
    html = html_template.format(origin=url, content=content)
    
    filename =(str(index) + '.' + str(title) + ".html").replace(':', '_')
    with open(filename, "w") as f:
        f.write(html.replace(u'\xa0', u' ').replace(u'\U0001f60a', u' ').replace(u'\u200b', u' '))
    print('完成', index)
    index += 1
#     break
print('## 全部完成 ##')
```
