```python
url = 'http://keras-cn.readthedocs.io/en/latest/'
```

```python
from urllib import request
req = request.Request(url)
res = request.urlopen(req)
html = res.read().decode('utf-8')
print(html)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
alist = soup.select('.wy-menu')[0].select('a')
alist[0:5]
```

```python
items = []
for item in alist:
    if  '#' not in item.get('href'):
        items.append('http://keras-cn.readthedocs.io/en/latest/' + item.get('href'))
items
```

```python
import hashlib
def md5(name):
    """ å°†å­—ç¬¦ä¸²è½¬æˆå“ˆå¸Œç  """
    if not isinstance(name, str):
        name = str(name)
    md5 = hashlib.md5()
    md5.update(name.encode('utf-8'))
    return md5.hexdigest()
```

```python
from bs4 import BeautifulSoup
import re
import os
pattern = '<img .*?src=\"(.*?)\"'
re_image = re.compile(pattern)
index = 1
for item in items:
    res = request.urlopen(item)
    html = res.read().decode('utf-8')
    
    soup = BeautifulSoup(html, 'html.parser')
    content = soup.select('.section')[0]
    title = soup.select('.section')[0].h1.text.strip()
    
    content = str(content)

    for image_link in re_image.findall(content):
        if not os.path.exists('images'):
            os.mkdir('images')
        image_link2 = 'http://keras-cn.readthedocs.io/en/latest/' + image_link.lstrip('./')
        filename = 'images/' + md5(image_link2) + os.path.splitext(image_link2)[-1]
        try:
            request.urlretrieve(image_link2, filename)
            print('ä¸‹è½½å®Œæˆ', filename)
        except Exception as e:
            print('å›¾ç‰‡å‡ºé”™', e)
        else:
            content = content.replace(image_link, filename)
    print('å·²å®Œæˆ--------')
    
    html_template = """<!DOCTYPE html>
    <html><head><meta charset="UTF-8">
    </head><body>
    <p><a href="{origin}">åŸæ–‡é“¾æ¥</a></p>
        {content}
    </body></html>"""
    html = html_template.format(origin=url, content=content)
    
    filename =(str(index) + '.' + str(title) + ".html").replace(':', '_')
    with open(filename, "w") as f:
        f.write(html.replace(u'\xa0', u' ').replace(u'\U0001f60a', u' ').replace(u'\u200b', u' '))
    print('å®Œæˆ', index)
    index += 1
#     break
```
