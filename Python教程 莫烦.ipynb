```python
# url = 'https://morvanzhou.github.io/tutorials/machine-learning/keras/'  # Keras
# url = 'https://morvanzhou.github.io/tutorials/machine-learning/tensorflow/'  # TensorFlow
url = 'https://morvanzhou.github.io/tutorials/machine-learning/sklearn/'  # Sciki-learn
```

```python
from urllib import request
req = request.Request(url)
res = request.urlopen(req)
html = res.read().decode('utf-8')
print(html)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html, 'html.parser')
alist = soup.select('.table-of-contents-ul')[0].select('a')
alist[0:5]
```

```python
items = []
for item in alist:
    if  '#' not in item.get('href'):
        items.append('https://morvanzhou.github.io' + item.get('href'))
items[:5]
```

```python
import hashlib
def md5(name):
    """ 将字符串转成哈希码 """
    if not isinstance(name, str):
        name = str(name)
    md5 = hashlib.md5()
    md5.update(name.encode('utf-8'))
    return md5.hexdigest()
```

```python
from bs4 import BeautifulSoup
import re
import os
pattern = '<img .*?src=\"(.*?)\"'
re_image = re.compile(pattern)
index = 1
for item in items:
    res = request.urlopen(item)
    html = res.read().decode('utf-8')
    
    soup = BeautifulSoup(html, 'html.parser')
    content = soup.select('.tut-main-content-pad')[0]
    title = soup.select('.tut-main-content-pad')[0].h1.text.strip()
    
    content = str(content)

    for image_link in re_image.findall(content):
        if not os.path.exists('images'):
            os.mkdir('images')
        image_link2 = 'https://morvanzhou.github.io' + image_link
        filename = 'images/' + md5(image_link2) + os.path.splitext(image_link2)[-1]
        try:
            request.urlretrieve(image_link2, filename)
            print('下载完成', filename)
        except Exception as e:
            print('图片出错', e)
        else:
            content = content.replace(image_link, filename)
    print('已完成--------')
    
    html_template = """<!DOCTYPE html>
    <html><head><meta charset="UTF-8">
    </head><body>
    <p><a href="{origin}">原文链接</a></p>
        {content}
    </body></html>"""
    html = html_template.format(origin=url, content=content)
    
    filename =(str(index) + '.' + str(title) + ".html").replace(':', '_')
    with open(filename, "w") as f:
#         f.write(html.replace(u'\xa0', u' ').replace(u'\U0001f60a', u' ').replace(u'\u200b', u' '))
        f.write(html)
    print('当前文件完成', index)
    index += 1
#     break
print('## 全部完成 ##')
```
